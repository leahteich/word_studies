```python
import pandas as pd
import numpy as np
```

```python
len_data = pd.read_csv("length_df_final.csv")
syll_data = pd.read_csv("syl_df_data.csv")
```

```python
syll_data
```

```python
def commonality_len(word1_len, word2_len):
    commonality = []
    n = 0

    for index, row in len_data.iterrows():
        if row['Word1_Length'] == word1_len: 
            if row['Word2_Length'] == word2_len:
                commonality.append(row["Freq_Per_Mil2"]/row["Freq_Per_Mil1"])
                n+=1 

    return("a word of length " + str(word2_len) + " is on avg " + str(sum(commonality)/len(commonality)) + "x more frequent than a word of length " +str(word1_len) + " for " + str(n) +" pairs")
```

```python
print(commonality_len(4,3))
print(commonality_len(5,3))
print(commonality_len(6,3))
print(commonality_len(7,3))
print(commonality_len(8,3))

print(commonality_len(5,4))
print(commonality_len(6,4))
print(commonality_len(7,4))
print(commonality_len(8,4))

print(commonality_len(6,5))
print(commonality_len(7,5))
print(commonality_len(8,5))

print(commonality_len(7,6))
print(commonality_len(8,6))

print(commonality_len(8,7))
```
